```python
import pandas as pd

# Load Silver Layer
weather = pd.read_parquet("weather_cleaned.parquet")
traffic = pd.read_parquet("traffic_cleaned.parquet")

print("Silver data loaded ")
```

```python
# Convert date_time column to proper datetime
weather['date_time'] = pd.to_datetime(weather['date_time'])
traffic['date_time'] = pd.to_datetime(traffic['date_time'])

# Create new time-based features (day, hour)
weather['day'] = weather['date_time'].dt.date
weather['hour'] = weather['date_time'].dt.hour

# Feature Engineering for Gold Layer
traffic['day'] = traffic['date_time'].dt.date
traffic['hour'] = traffic['date_time'].dt.hour
```

```python
# Create Daily Aggregations
# Summaries used for dashboards and analytics
weather_daily = weather.groupby("day").mean(numeric_only=True)
traffic_daily = traffic.groupby("day").mean(numeric_only=True)
```

```python
weather.to_parquet("weather_gold.parquet")
traffic.to_parquet("traffic_gold.parquet")

weather_daily.to_parquet("weather_daily.parquet")
traffic_daily.to_parquet("traffic_daily.parquet")

print("Gold files created successfully ")
```
